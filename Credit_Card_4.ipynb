```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

import warnings
warnings.filterwarnings("ignore")
```

```python
da=pd.read_csv('application_record.csv')
da
```

```python
dc=pd.read_csv('credit_record.csv')
dc
```

```python
dc['STATUS'].value_counts()
```

```python
dc['STATUS'].replace(['1','2','3','4','5'],1,inplace=True)
dc['STATUS'].value_counts()
```

```python
dc['STATUS'].replace(['C','X','0'],0,inplace=True)
dc['STATUS'].value_counts()
```

```python
type(good[6])
```

```python
dc=dc.groupby('ID').agg({
                        'STATUS':sum
                        ,'MONTHS_BALANCE': 'count'
})
```

```python
dc
```

```python
for i in dc.index.tolist():
        if dc.loc[i,'STATUS']==0:
            dc.loc[i,'STATUS']=1
        else:
            dc.loc[i,'STATUS']=0
```

```python
dc['STATUS'].value_counts()
```

* ## merging the two data sets on column ID to not Coflict the data 
***

```python
data = pd.merge(da, dc, on="ID",how='inner')
data
```

* ## replace the opject values to numirec
****

```python
data['CODE_GENDER'].replace(['F','M'], [0,1],inplace=True)
data['FLAG_OWN_CAR'].replace(['Y','N'], [1,0],inplace=True)
data['FLAG_OWN_REALTY'].replace(['Y','N'], [1,0],inplace=True)
```

```python
# drop column 'FLAG_MOBIL' cuse of the fixed value
data.drop(columns=['FLAG_MOBIL'],inplace =True)
```

```python
data.drop('ID',inplace=True,axis=1)
data
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isna().sum()
```

```python
data[data['OCCUPATION_TYPE'].isna()]['NAME_EDUCATION_TYPE'].value_counts()
```

```python
data[data['NAME_EDUCATION_TYPE']=='Secondary / secondary special']['OCCUPATION_TYPE'].value_counts()
```

```python
data['OCCUPATION_TYPE'].fillna('Laborers',inplace=True)
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

* ### split the data into train and test 
***

```python
x=data.drop(['STATUS'],axis=1).values
y=data.loc[:,'STATUS'].values
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3)
```

## Over Sampling

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
print ('data befor OverSampling ',Counter(train_y))

sm=SMOTE()
x_res,y_res=sm.fit_resample(train_x,train_y)
print ('data After OverSmpling ',Counter(y_res))
```

## Under Sampling

```python
from imblearn.under_sampling import TomekLinks
from collections import Counter
print ('data Befor UnderSampling ',Counter(y_res))
tl=TomekLinks(sampling_strategy='all')
x_res,y_res=tl.fit_resample(x_res,y_res)
print('data After underSampling',Counter(y_res)) 
```

* ### applaing the machine learning model Classification 
***

```python
from sklearn.preprocessing import StandardScaler

sd= StandardScaler()
train_x=sd.fit_transform(train_x)
test_x=sd.transform(test_x)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=40,max_depth=45,class_weight='balanced',random_state=60)
model.fit(train_x,train_y)
```

```python
y_pred=model.predict(test_x)
from sklearn.metrics import confusion_matrix
print (confusion_matrix(y_pred,test_y))
```

```python
# ACC on test
y_pred=model.predict(test_x)
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score
print ('Accuracy : ', accuracy_score(test_y,y_pred))
print ('Recall : ',recall_score(test_y,y_pred))
print ('precision : ',precision_score(test_y,y_pred))
print ('F1Score : ',f1_score(test_y,y_pred))
```

```python
# ACC on train
y_pred=model.predict(train_x)
print ('Accuracy : ', accuracy_score(train_y,y_pred))
print ('Recall : ',recall_score(train_y,y_pred))
print ('precision : ',precision_score(train_y,y_pred))
print ('F1Score : ',f1_score(train_y,y_pred))
```

```python
y_pred=model.predict(train_x)
cm=confusion_matrix(y_pred,train_y)
```

```python
specifict=cm[0,0]/(cm[0,0]+cm[0,1])
print('specifict : ',specifict)
```
